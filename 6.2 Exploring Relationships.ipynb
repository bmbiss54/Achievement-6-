```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

Data points closer to beige are more closely correlated and closer to dark purple shows poor correlation.

There are a number of correlations in the data set. It shows there is small correlation between price and certain features of the listing by accommodation, bedrooms, and beds. The calculated host listing is closely tied with the host total listing count which is expected as they are linked and similar data points. Lastly the number of reviews and total days the listing is available throughout the year shows a high correlations. 

```python
# Create a scatterplot for the "accommodates" and "price" columns in seaborn

sns.lmplot(x = 'accommodates', y = 'price', data = df_correct)
```

Here the price and amount of accommodations does show a decent correlationa and positive linear line. Here as you add accommodations the average price of the listing increases.
